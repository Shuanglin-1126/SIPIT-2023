```python
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt

img = io.imread(r'F:\毕设\实验\bianyaqi2\xisuhe\right_top\01.bmp')
#高斯滤波器
blur3 = cv2.GaussianBlur(img, (5, 5), 0)
#双边滤波器，同时考虑像素的空间位置和邻域的像素灰度相似性，在滤波的同时较好的保留图片的边缘
blur4 = cv2.bilateralFilter(img, 9, 75, 75)

plt.figure(figsize=(10, 10))
plt.subplot(221), plt.imshow(img), plt.title('img')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(blur3), plt.title('blur3')
plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.imshow(blur4), plt.title('blur4')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#将图像从RGB颜色空间转到HSV颜色空间
hsv = cv2.cvtColor(blur4,cv2.COLOR_BGR2HSV)
```

```python
# 颜色阈值分割
low_blue = np.array([55,0,0])
high_blue = np.array([118,255,255])
mask = cv2.inRange(hsv,low_blue,high_blue)
cv2.imshow('mask',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
